```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0,6])
y = np.array([0,80])
```

```python
plt.plot(x,y)
plt.show()
```

Line Chart

```python
points = np.array([5,7,12,14,4,12])
plt.plot(points)
plt.show()
```

Linestyle

```python
a2 = np.array([3,7,9,4,7,8,12])
plt.plot(a2,linestyle ="dashed")
plt.show()
```

Bar Chart

```python
x = np.array(["W","X","Y","Z"])
y = np.array([23,27,24,21])
plt.bar(x,y)
plt.show()
```

Scatter Plot

```python
x1 = np.array([20,16,12,10,17,19])
y1 = np.array([8,12,10,14,9,5])
plt.scatter(x1,y1)
plt.show()
```

Pie Chart

```python
p = np.array([30,10,20,40,26])
plt.pie(p)
```

Histogram

```python
h1 = np.random.normal(170,30,230)
plt.hist(h1)
```
